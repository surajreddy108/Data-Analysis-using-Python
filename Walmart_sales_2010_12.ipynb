```python
import pandas as pd
df=pd.read_csv('/content/Walmart Non Holiday Weekly Sales.csv')
```

```python
df
```

```python
df=df.drop(columns=['year'])
```

```python
df
```

**Finding year from the date**

```python
df.dtypes
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.dtypes
```

```python
df['Year']=df['Date'].dt.year
df
```

**Add a new column "Month" as the second column**

```python
df['Month']=df['Date'].dt.month
df
```

```python
df.insert(2,'Month',df.pop('Month'))
#df.insert(index_number,'Name of Col',col_values)
```

```python
df
```

**Top 10 days having maximum Weekly sales**

```python
df.sort_values('Weekly_Sales',ascending=False).head(10)
```

**Maximum and Minimum sales data for a week**

```python
df.groupby('week').Weekly_Sales.max().sort_values(ascending=False).head(1)
```

```python
df.groupby('week').Weekly_Sales.min().sort_values(ascending=True).head(1)
```
